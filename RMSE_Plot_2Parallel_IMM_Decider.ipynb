```python
# This code plot RMSEs for different tuning parameters.


import numpy as np
import matplotlib.pyplot as plt
import math
import pylab

#x_tune_set = 6;
#y_tune_set = 6;
#o_tune_set = 4;

pi = math.pi

#===========================================================================IMM Default and High Perf

Data_file_D_H = np.loadtxt('Output_IMM4_Default_HighPerf_0.txt' ,delimiter=',')

d_0 = Data_file_D_H.shape
d1_0 = np.asarray(d_0)
d2_0 = d1_0[0]
d3_0 = d1_0[1]

d3 = d3_0

#================================================= For IMM High perf
d2 = int(d2_0/2)

Data_file = np.zeros(shape=(d2,d3_0))

for i in range(d2):
    
    j = int(2*i)
    Data_file[i,:] = Data_file_D_H[j,:]
    
#-------------------
Data_file_1 = np.zeros(shape=(d2,d3_0))

for i in range(d2):
    
    j = int(2*i+1)
    Data_file_1[i,:] = Data_file_D_H[j,:]
    

#==============================================
        
t = Data_file[:,18]
t = t.reshape(1,d2)

#===========================================RMSE

Meas = np.zeros(shape=(d2,d3))
Meas_1 = np.zeros(shape=(d2,d3))

pos_MSE = np.zeros(shape=(1,d2))
spd_smooth_MSE = np.zeros(shape=(1,d2))
#----------
spd_filter_MSE = np.zeros(shape=(1,d2))
#----------
head_smooth_MSE = np.zeros(shape=(1,d2))
#----------
head_filter_MSE = np.zeros(shape=(1,d2))
#----------
pos_MSE_1 = np.zeros(shape=(1,d2))
spd_smooth_MSE_1 = np.zeros(shape=(1,d2))
#----------
spd_filter_MSE_1 = np.zeros(shape=(1,d2))
#----------
head_smooth_MSE_1 = np.zeros(shape=(1,d2))
#----------
head_filter_MSE_1 = np.zeros(shape=(1,d2))
#----------
pos_MSE_dec = np.zeros(shape=(1,d2))
spd_smooth_MSE_dec = np.zeros(shape=(1,d2))
#----------
spd_filter_MSE_dec = np.zeros(shape=(1,d2))
#----------
head_smooth_MSE_dec = np.zeros(shape=(1,d2))
#----------
head_filter_MSE_dec = np.zeros(shape=(1,d2))
#----------

#turn_MSE = np.zeros(shape=(1,d2))
#--------------
#turn_sm = np.zeros(shape=(1,d2))
#accel_sm = np.zeros(shape=(1,d2))
#--------------
#accel_MSE = np.zeros(shape=(1,d2))
#prob_ss = np.zeros(shape=(1,d2))
#prob_turn = np.zeros(shape=(1,d2))
#prob_highG = np.zeros(shape=(1,d2))
#prob_accel = np.zeros(shape=(1,d2))
        

MC = 100


#----------------------------- Decider index

D_i_1 = np.zeros(shape=(1,d2))

D_i = np.zeros(shape=(1,d2))

#-----------------------------


for i in range(MC):
    
    #i = i+29   #********************************* delete it
    
    Meas_D_H = np.loadtxt('Output_IMM4_Default_HighPerf_' + str(i) + '.txt' ,delimiter=',')
    
    #Meas = np.loadtxt('Output_IMM4_4.txt' ,delimiter=',')
    
    #================================================= For IMM High perf


    for ii in range(d2):
    
        Meas[ii,:] = Meas_D_H[2*ii+1,:]
        Meas_1[ii,:] = Meas_D_H[2*ii,:]
    

    #================================================

        
        #------------------------------------------High Perf IMM
        
        temp_x_sm = Meas[ii,8]
        #temp_x_sm = temp.reshape(1,d2)
    
        temp_x_tru = Meas[ii,19]
        #temp_x_tru = temp.reshape(1,d2)
    
        temp_y_sm = Meas[ii,9]
        #temp_y_sm = temp.reshape(1,d2)
    
        temp_y_tru = Meas[ii,20]
        #temp_y_tru = temp.reshape(1,d2)
    
        temp_spd_sm = Meas[ii,13]
        #temp_spd_sm = temp.reshape(1,d2)
    
        #-------------------
        temp_spd_fl = Meas[ii,14]
        #temp_spd_fl = temp.reshape(1,d2)
        #-------------------
    
        temp_spd_tru = Meas[ii,22]*3600
        #temp_spd_tru = temp.reshape(1,d2)
    
        temp_head_sm = Meas[ii,15]*180/pi
        #temp_head_sm = temp.reshape(1,d2)
    
        #-------------------
        temp_head_fl = Meas[ii,16]
        #temp_head_fl = temp.reshape(1,d2)
        #-------------------
    
        temp_head_tru = Meas[ii,23]*180/pi
        #temp_head_tru = temp.reshape(1,d2)
        
        temp_accel_tru = Meas[ii,26]*0.51444*3600
       #temp_accel_tru = temp.reshape(1,d2)
    
        temp_accel_sm = Meas[ii,27]*0.51444*3600
       #temp_accel_sm = temp.reshape(1,d2)
    
        temp_turn_tru = Meas[ii,25]*180/pi
       #temp_turn_tru = temp.reshape(1,d2)
    
        temp_turn_sm = Meas[ii,28]*180/pi
       #temp_turn_sm = temp.reshape(1,d2)
    
       #temp = Meas[:,29]
       #temp_prob_ss = temp.reshape(1,d2)
    
       #temp = Meas[:,30]
       #temp_prob_turn = temp.reshape(1,d2)    
    
       #temp = Meas[:,31]
       #temp_prob_highG = temp.reshape(1,d2)
    
       #temp = Meas[:,32]
       #temp_prob_accel = temp.reshape(1,d2)
    
    #--------------------------------------------------
    
    #-------------------------------------------------- Default IMM
    
        temp_x_sm_1 = Meas_1[ii,8]
        #temp_x_sm = temp.reshape(1,d2)
    
        temp_x_tru_1 = Meas_1[ii,19]
        #temp_x_tru = temp.reshape(1,d2)
    
        temp_y_sm_1 = Meas_1[ii,9]
        #temp_y_sm = temp.reshape(1,d2)
    
        temp_y_tru_1 = Meas_1[ii,20]
        #temp_y_tru = temp.reshape(1,d2)
    
        temp_spd_sm_1 = Meas_1[ii,13]
        #temp_spd_sm = temp.reshape(1,d2)
    
        #-------------------
        temp_spd_fl_1 = Meas_1[ii,14]
        #temp_spd_fl = temp.reshape(1,d2)
        #-------------------
    
        temp_spd_tru_1 = Meas_1[ii,22]*3600
        #temp_spd_tru = temp.reshape(1,d2)
    
        temp_head_sm_1 = Meas_1[ii,15]
        #temp_head_sm = temp.reshape(1,d2)
    
        #-------------------
        temp_head_fl_1 = Meas_1[ii,16]
        #temp_head_fl = temp.reshape(1,d2)
        #-------------------
    
        temp_head_tru_1 = Meas_1[ii,23]*180/pi
        #temp_head_tru = temp.reshape(1,d2)
        
        temp_accel_tru_1 = Meas_1[ii,26]*0.51444*3600
       #temp_accel_tru = temp.reshape(1,d2)
    
        temp_accel_sm_1 = Meas_1[ii,27]*0.51444*3600
       #temp_accel_sm = temp.reshape(1,d2)
    
        temp_turn_tru_1 = Meas_1[ii,25]*180/pi
       #temp_turn_tru = temp.reshape(1,d2)
    
        temp_turn_sm_1 = Meas_1[ii,28]*180/pi
       #temp_turn_sm = temp.reshape(1,d2)
    
       #temp = Meas[:,29]
       #temp_prob_ss = temp.reshape(1,d2)
    
       #temp = Meas[:,30]
       #temp_prob_turn = temp.reshape(1,d2)    
    
       #temp = Meas[:,31]
       #temp_prob_highG = temp.reshape(1,d2)
    
       #temp = Meas[:,32]
       #temp_prob_accel = temp.reshape(1,d2)
    
    #--------------------------------------------------
    
            
        while temp_head_tru>360:
            temp_head_tru = temp_head_tru - 360
            
        while temp_head_tru<0:
            temp_head_tru = temp_head_tru + 360
            
            
        while temp_head_tru_1>360:
            temp_head_tru_1 = temp_head_tru_1 - 360
            
        while temp_head_tru_1<0:
            temp_head_tru_1 = temp_head_tru_1 + 360
            
        #temp = temp.reshape(d2,1)
                                
                    
        #------------------------------------------ IMM Default
            
        
        pos_Error_1 = (temp_x_sm_1 - temp_x_tru_1)**2 + (temp_y_sm_1-temp_y_tru_1)**2
        pos_MSE_1[0,ii] = pos_MSE_1[0,ii] + pos_Error_1
        
        spd_fl_Error_1 = (temp_spd_fl_1-temp_spd_tru_1)**2
        spd_filter_MSE_1[0,ii] = spd_filter_MSE_1[0,ii] + spd_fl_Error_1
        
        spd_sm_Error_1 = (temp_spd_sm_1-temp_spd_tru_1)**2
        spd_smooth_MSE_1[0,ii] = spd_smooth_MSE_1[0,ii] + spd_sm_Error_1
        
    #head_MSE = head_MSE + (temp_head_sm[0,:]-temp_head_tru[0,:])**2
    
    
        Err_sm_abs_1 = np.abs(temp_head_sm_1-temp_head_tru_1)
        Err_fl_abs_1 = np.abs(temp_head_fl_1-temp_head_tru_1)
        
        if (Err_sm_abs_1>180):
        
            Err_sm_abs_1 = Err_sm_abs_1 - 360
            
        head_sm_Error_1 = (Err_sm_abs_1)**2    
        head_smooth_MSE_1[0,ii] = head_smooth_MSE_1[0,ii] + head_sm_Error_1
        
        if (Err_fl_abs_1>180):
        
            Err_fl_abs_1 = Err_fl_abs_1 - 360
        
        head_fl_Error_1 = (Err_fl_abs_1)**2
        head_filter_MSE_1[0,ii] = head_filter_MSE_1[0,ii] + head_fl_Error_1
        
        #------------------------------------------- IMM High Perf
         
        pos_Error = (temp_x_sm - temp_x_tru)**2 + (temp_y_sm-temp_y_tru)**2
        pos_MSE[0,ii] = pos_MSE[0,ii] + pos_Error
        
        spd_fl_Error = (temp_spd_fl-temp_spd_tru)**2
        spd_filter_MSE[0,ii] = spd_filter_MSE[0,ii] + spd_fl_Error
        
        spd_sm_Error = (temp_spd_sm-temp_spd_tru)**2
        spd_smooth_MSE[0,ii] = spd_smooth_MSE[0,ii] + spd_sm_Error
        
    #head_MSE = head_MSE + (temp_head_sm[0,:]-temp_head_tru[0,:])**2
    
    
        Err_sm_abs = np.abs(temp_head_sm-temp_head_tru)
        Err_fl_abs = np.abs(temp_head_fl-temp_head_tru)
        
        if (Err_sm_abs>180):
        
            Err_sm_abs = Err_sm_abs - 360
            
        head_sm_Error = (Err_sm_abs)**2    
        head_smooth_MSE[0,ii] = head_smooth_MSE[0,ii] + head_sm_Error
        
        if (Err_fl_abs>180):
        
            Err_fl_abs = Err_fl_abs - 360
        
        head_fl_Error = (Err_fl_abs)**2
        head_filter_MSE[0,ii] = head_filter_MSE[0,ii] + head_fl_Error
            
           
        if np.abs(temp_spd_sm)<500 and np.abs(temp_turn_sm)<4 and np.abs(temp_spd_sm_1)<500 and np.abs(temp_turn_sm_1)<4:
            
            # The arguments in the above if statement can be studied further
            
            D_i_1[0,ii] = D_i_1[0,ii] + 1
                
                
            pos_MSE_dec[0,ii] = pos_MSE_dec[0,ii] + pos_Error_1
            
            spd_smooth_MSE_dec[0,ii] = spd_smooth_MSE_dec[0,ii] + spd_sm_Error_1
        
            spd_filter_MSE_dec[0,ii] = spd_filter_MSE_dec[0,ii] + spd_fl_Error_1
            
            head_smooth_MSE_dec[0,ii] = head_smooth_MSE_dec[0,ii] + head_sm_Error_1
            
            head_filter_MSE_dec[0,ii] = head_filter_MSE_dec[0,ii] + head_fl_Error_1
                    
            
        else:
            
            D_i[0,ii] = D_i[0,ii] + 1
            
            pos_MSE_dec[0,ii] = pos_MSE_dec[0,ii] + pos_Error
            
            spd_smooth_MSE_dec[0,ii] = spd_smooth_MSE_dec[0,ii] + spd_sm_Error
        
            spd_filter_MSE_dec[0,ii] = spd_filter_MSE_dec[0,ii] + spd_fl_Error
            
            head_smooth_MSE_dec[0,ii] = head_smooth_MSE_dec[0,ii] + head_sm_Error
            
            head_filter_MSE_dec[0,ii] = head_filter_MSE_dec[0,ii] + head_fl_Error
            
            
            
            
            
            
pos_RMSE = np.sqrt(pos_MSE/MC)
spd_smooth_RMSE = np.sqrt(spd_smooth_MSE/MC)
spd_filter_RMSE = np.sqrt(spd_filter_MSE/MC)
head_smooth_RMSE = np.sqrt(head_smooth_MSE/MC)
head_filter_RMSE = np.sqrt(head_filter_MSE/MC)


pos_RMSE_1 = np.sqrt(pos_MSE_1/MC)
spd_smooth_RMSE_1 = np.sqrt(spd_smooth_MSE_1/MC)
spd_filter_RMSE_1 = np.sqrt(spd_filter_MSE_1/MC)
head_smooth_RMSE_1 = np.sqrt(head_smooth_MSE_1/MC)
head_filter_RMSE_1 = np.sqrt(head_filter_MSE_1/MC)

pos_RMSE_dec = np.sqrt(pos_MSE_dec/MC)
spd_smooth_RMSE_dec = np.sqrt(spd_smooth_MSE_dec/MC)
spd_filter_RMSE_dec = np.sqrt(spd_filter_MSE_dec/MC)
head_smooth_RMSE_dec = np.sqrt(head_smooth_MSE_dec/MC)
head_filter_RMSE_dec = np.sqrt(head_filter_MSE_dec/MC)


#================================================================================End of IMM High Perf
```
